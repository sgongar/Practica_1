```python
from os import getcwd

import matplotlib.pyplot as plt
from numpy import arange, log10, nan, linspace
import pandas as pd

import networkx as nx
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(11.69, 8.27))

#
# Obtenemos la masa y edad de la principal
# 
total_df = pd.read_csv(getcwd() + '/datos_completos/galaxias_13.csv')

z_list = []
y_list = []
xz_list = []
size_list = []
colour_list = []
age_list = []
x_list = []
x = 0
for idx, row in total_df.iterrows():
    if row['stellarMass'] > 0.1:
        print(row['galaxyID'])
        x_list.append(x)
        x = x + 0.1
        z_list.append(float(row['z']))
        y_list.append(float(row['y']))
        xz_list.append(float(row['z'])*float(row['x']))
        age_list.append(float(row['massWeightedAge']))
        size_list.append(float(row['stellarMass'])*7)
        colour_list.append(float(row['mag_b'])-float(row['mag_v']))

max_b_v = max(total_df['mag_b']-total_df['mag_v'])
min_b_v = min(total_df['mag_b']-total_df['mag_v'])

t = linspace(start=min_b_v, stop=max_b_v, num=len(z_list))

ax.scatter(x_list, age_list, s=size_list, c=colour_list, cmap='rainbow')

plt.arrow(0, -0.00010, 10, -0.00005, length_includes_head=True,
          head_width=0.08, head_length=0.00002)

# Muestra por pantalla
plt.show()
# Guarda la figura
# plt.savefig('final_2.png')
```
